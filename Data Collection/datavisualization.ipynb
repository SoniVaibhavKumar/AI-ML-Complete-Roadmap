```python
# Real data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M

# Plot
plt.plot(years, oscar_revenue)
plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

plt.plot(years, oscar_revenue, color="red", marker="o", linestyle="-", linewidth=5, label="Oscar Movies")
plt.plot(years, non_oscar_revenue, label="Non-Oscar Movies")

plt.title("Oscar Movies vs Non-Oscar movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")

plt.legend()
```

```python
import numpy as np

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M

non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

years = [2008, 2009, 2010, 2011, 2012]

x = np.arange(len(years))    
width = 0.25

plt.bar(x-width, oscar_revenue, width, label="Oscar Movies")
plt.bar(x, non_oscar_revenue, width, label="Non-Oscar Movies")
plt.bar(x+width, oscar_revenue, width, label="Oscar Movies")

plt.title("Oscar vs Non-Oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue($M)")
plt.legend()
```

```python
# Scatter Plots

# Scatter Plots
people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

# plot
# plt.scatter(age, blood_pressure, color="red")

plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()

# Customizing plots
colors = ["green" if bp<135 else "red" for bp in blood_pressure]
sizes = blood_pressure
plt.scatter(age, blood_pressure, color=colors, s=sizes, marker='^', alpha=0.5)

# plt.scatter(age, blood_pressure, c=blood_pressure, cmap="plasma")
plt.scatter(age, blood_pressure, c=blood_pressure, cmap="OrRd")# makes more sense
# plt.scatter(age, blood_pressure, c=[5]*len(age), cmap="plasma")
 plt.colorbar(label="BP")

for i in range(len(age)):
    plt.annotate(people[i], xy=(age[i], blood_pressure[i]))
plt.xlim(min(age), max(age)+10)
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

# colors = ["red", "black", "pink", "blue", "yellow"]
# plt.pie(amounts, labels=expenses, autopct="%1.1f%%", colors=colors)

explode = [0, 0, 0, 0.2, 0]

# plt.style.use("classic")
plt.style.use("default")
plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
       wedgeprops = {
           "edgecolor" : "black",
           "linewidth" : 2,
           "linestyle" : "--",
           
       },
        explode=explode,
        startangle=360,
        shadow=True
       )
plt.show()
```
